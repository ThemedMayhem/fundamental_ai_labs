```python
def init_params():
    Whh = np.random.rand(100, 100)
    Wyh = np.random.rand(100, 1)
    Wxh = np.random.rand(5,100)
    return Whh, Wyh, Wxh

def tanh(x):
    return np.tanh(x)

def loss_mse(y,y_pred):
    return sum(((y-y_pred)**2))

def loss_rmse(y,y_pred):
    return np.sqrt(sum(((y-y_pred)**2)))

def loss_r2(y,y_pred):
    y_mean = np.mean(y)
    return 1 - sum(((y-y_pred)**2))/(sum((y-y_mean)**2)+0.000000001)
    
def forward_prop(Whh, Wyh, Wxh, X):
    y = np.zeros((len(X)))
    Xh0 = Wxh.T.dot(X[0])
    H0 = tanh(Xh0)
    y[0] = Wyh.T.dot(H0)
    
    Xh1 = Whh.dot(H0) + Wxh.T.dot(X[1])
    H1 = tanh(Xh1)
    y[1] = Wyh.T.dot(H1)
    
    Xh2 = Whh.dot(H1) + Wxh.T.dot(X[2])
    H2 = tanh(Xh2)
    y[2] = Wyh.T.dot(H2)  
    
    return Xh0,Xh1,Xh2,H0,H1,H2,y



def backward_prop(Whh, Wyh, Wxh, Xh0,Xh1,Xh2,H0,H1,H2,y,X, Y):
    m = len(Y)
    dE = ((y - Y).T)
    X = X.reshape(3,1,5)
    Xh2 = Xh2.reshape(100,1)
    Xh1 = Xh1.reshape(100,1)
    Xh0 = Xh0.reshape(100,1)
    dWyh = dE[0]*H0 + dE[1]*H1 + dE[2]*H2
    
    dWxh = (dE[2]*Wyh.T.dot((1-(tanh(Xh2))**2).dot(X[2]))) + (dE[2]*Wyh.T.dot(1-(tanh(Xh2))**2)[0][0]*(Whh).dot(1-(tanh(Xh1))**2).dot(X[1]) + dE[1]*Wyh.T.dot(1-(tanh(Xh1))**2).dot(X[1])) + \
    (dE[2]*Wyh.T.dot(1-(tanh(Xh2))**2)[0][0]*(Whh).dot(1-(tanh(Xh1))**2).T.dot(Whh).dot(1-(tanh(Xh0))**2).dot(X[0])  +  dE[1]*Wyh.T.dot(1-(tanh(Xh1))**2)[0][0]*(Whh).dot(1-(tanh(Xh0))**2).dot(X[0]) + dE[0]*Wyh.T.dot(1-(tanh(Xh0))**2).dot(X[0])       )
    
    
    dWhh = (dE[2]*Wyh*(1-(tanh(Xh2))**2)*H1) + (dE[2]*Wyh*(1-(tanh(Xh2))**2)*Whh*(1-(tanh(Xh1))**2)*H0) + (dE[1]*Wyh*(1-(tanh(Xh1))**2)*H0)
    
    return dWyh, dWxh, dWhh

def update_params(Whh, Wyh, Wxh, dWyh, dWxh, dWhh, alpha):
    dWyh = dWyh.reshape((100,1))
    Whh = Whh - alpha * dWhh   
    Wyh = Wyh - alpha * dWyh
    Wxh = Wxh - alpha * dWxh.T 
    return Whh, Wyh, Wxh
```

```python
def init_params():
    Wfx = np.random.rand(5, 5)
    Wrx = np.random.rand(5, 5)
    Wfh = np.random.rand(5, 5)
    Wrh = np.random.rand(5, 5)
    Wy = np.random.rand(1, 5)
    return Wfx,Wrx,Wfh,Wrh,Wy

def tanh(x):
    return np.tanh(x)

def sigmoid(x):
    return (1/(1 + np.exp(-x)))


def loss_mse(y,y_pred):
    return sum(((y-y_pred)**2))

def loss_rmse(y,y_pred):
    return np.sqrt(sum(((y-y_pred)**2)))

def loss_r2(y,y_pred):
    y_mean = np.mean(y)
    return (1 - (sum(((y-y_pred)**2))/(sum((y-y_mean)**2)+0.0000000000000000000000000000001)))
    
def forward_prop(Wfx,Wrx,Wfh,Wrh,Wy, X):
    
    h = np.zeros((len(X),5))
    y = np.zeros((len(X)))
    f0 = sigmoid(Wfx.dot(X[0]))
    R0 = tanh(Wrx.dot(X[0]))
    h[0] = f0*R0
    y[0] = Wy.dot(h[0])
    
    f1 = sigmoid(Wfx.dot(X[1]) + Wfh.dot(h[0]))
    R1 = tanh(Wrx.dot(X[1]) + Wrh.dot(f1*h[0]))
    h[0] = (1-f1)*h[0] + f1*R1
    y[1] = Wy.dot(h[1])
    
    return f0,R0,f1,R1,h,y



def backward_prop(f0,R0,f1,R1,Wfx,Wrx,Wfh,Wrh,Wy,h,y,X, Y):
    m = len(Y)
    dE = ((y - Y).T)
    f0 = f0.reshape((5,1))
    f1 = f1.reshape((5,1))
    R0 = R0.reshape((5,1))
    R1 = R1.reshape((5,1))

    
    dWrh = dE[1]*Wy.dot(f1).dot((1-tanh(Wrx.dot(X[0].reshape((5,1))))**2).T).dot(f1*h[0])
    
    dh0_dWrx = f0.T.dot(1-tanh(Wrx.dot(X[0].reshape((5,1))))**2)*(X[0].reshape((5,1)))
    df1_dWrx = (1-sigmoid(Wfx.dot(X[1].reshape((5,1))) + Wfh.dot(h[0]))).dot(sigmoid(Wfx.dot(X[1].reshape((5,1))) + Wfh.dot(h[0])))

    dWrx = dE[0]*Wy.dot(dh0_dWrx) + dE[1]*Wy.dot((1-f1)*(dh0_dWrx) - h[0]*(df1_dWrx)*(Wfh).dot(dh0_dWrx) + R1*(df1_dWrx) +\
                f1*((1-tanh(Wrx.dot(X[1].reshape((5,1)) + Wrh.dot(f1*h[0])))**2)).dot(X[1].reshape((5,1)) + Wrh.dot(df1_dWrx*(h[0]) + dh0_dWrx.dot(f1.T))))
   
    dh0_dWfx = R0.T.dot(1-sigmoid(Wfx.dot(X[0].reshape((5,1))))).dot(sigmoid(Wfx.dot(X[0].reshape((5,1)))).T).dot(X[0].reshape((5,1)))
    df1_dWfx = (1-sigmoid(Wfx.dot(X[1].reshape((5,1))) + Wfh.dot(h[0]))).dot(sigmoid(Wfx.dot(X[0].reshape((5,1))) + Wfh.dot(h[0])))*(R0).dot(dh0_dWfx)
    
    dWfx = dE[0]*Wy.T.dot(dh0_dWfx) + dE[1]*Wy.dot((1-f1).dot(dh0_dWfx) - h[0].dot(df1_dWfx) - f1.dot(dh0_dWfx) + R1.T.dot(df1_dWfx) + f1.T.dot((1-tanh(Wrx.dot(X[1] + Wrh.dot(f1*h[0])))**2))*(Wrh).dot(f1.dot(dh0_dWfx) + h[0].dot(df1_dWfx)))
   
   
    df1_dWfh =  (1-sigmoid(Wfx.dot(X[1].reshape((5,1))) + Wfh.dot(h[0]))).dot(sigmoid(Wfx.dot(X[0]) + Wfh.dot(h[0])))*(h[0])
    
    dWfh = dE[1]*Wy.dot(-h[0].dot(df1_dWfh) + R1.dot(df1_dWfh.reshape((5,1)).T) + f1.T.dot((1-tanh(Wrx.dot(X[1].reshape((5,1)) + Wrh.dot(f1*h[0])))**2)).dot(Wrh).dot(h[0])[0]*(df1_dWfh.reshape((5,1))))
    
   
    dWy = dE[0]*h[0] + dE[1]*h[1]                                                                
    
    return dWrh, dWrx, dWfx, dWfh, dWy

def update_params(Wfx,Wrx,Wfh,Wrh,Wy, dWrh, dWrx, dWfx, dWfh, dWy, alpha):
    Wrx = Wrx - alpha * dWrx
    Wfx = Wfx - alpha * dWfx   
    Wfh = Wfh - alpha * dWfh 
    Wrh = Wrh - alpha * dWrh
    Wy = Wy - alpha * dWy
    return Wfx,Wrx,Wfh,Wrh,Wy
```

```python
def gradient_descent(X, Y, alpha, iterations):
    print(len(X))
    Wfx,Wrx,Wfh,Wrh,Wy = init_params()
    n = 10
    for i in range(iterations):

        for o in range(10):
#             if j % 10 == 0: print(j)
            f0,R0,f1,R1,h,y = forward_prop(Wfx,Wrx,Wfh,Wrh,Wy, X[o:o+2])

            dWy, dWrh, dWrx, dWfx, dWfh = backward_prop(f0,R0,f1,R1,Wfx,Wrx,Wfh,Wrh,Wy,h,y,X[o:o+2], Y[o:o+2])
            Wfx,Wrx,Wfh,Wrh,Wy = update_params(Wfx,Wrx,Wfh,Wrh, Wy, dWrh, dWrx, dWfx, dWfh,dWy, alpha)
            

        if i % 1000 == 0:
            loss_mse_sum = 0
            loss_rmse_sum = 0
            loss_r2_sum = 0
            loss_mse_sum_val = 0
            loss_rmse_sum_val = 0
            loss_r2_sum_val = 0
            for j in range(10):
                _,_,_,_,_,y_pred = forward_prop(Wfx,Wrx,Wfh,Wrh,Wy, X[j:j+2])
                loss_mse_sum = loss_mse_sum + loss_mse(Y[j:j+2],y_pred)/2
                loss_rmse_sum = loss_rmse_sum + loss_rmse(Y[j:j+2],y_pred)/2
                loss_r2_sum = loss_r2_sum + loss_r2(Y[j:j+2],y_pred)/2
                _,_,_,_,_,y_pred = forward_prop(Wfx,Wrx,Wfh,Wrh,Wy, X[j+n:j+2+n])
                loss_mse_sum_val = loss_mse_sum_val + loss_mse(Y[j+n:j+n+2],y_pred)/2
                loss_rmse_sum_val = loss_rmse_sum_val + loss_rmse(Y[j+n:j+n+2],y_pred)/2
                loss_r2_sum_val = loss_r2_sum_val + loss_r2(Y[j+n:j+n+2],y_pred)/2
            print(f"epoch: [{i}/{iterations}], mse={np.round(loss_mse_sum,5)}, rmse={np.round(loss_rmse_sum,5)}, r2={np.round(loss_r2_sum,3)}")
                 # mse_val={np.round(loss_mse_sum_val,5)}, rmse_val={np.round(loss_rmse_sum_val,5)}, r2_val={np.round(loss_r2_sum_val,3)}         ")
            
    return Wfx,Wrx,Wfh,Wrh,Wy
```
